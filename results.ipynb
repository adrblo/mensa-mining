```python
%%sql

SELECT name_de, min(priceStudents), max(priceStudents), (max(priceStudents) / min(priceStudents) - 1) AS infl, min(date), max(date) FROM dish WHERE restaurant = 'mensa-academica-paderborn' AND category = 'dish' GROUP BY name_de ORDER BY infl DESC
```

```python
%%sql

SELECT ddish.name_de, start.priceStudents AS "earliest price", end.priceStudents AS "latest price", (end.priceStudents / start.priceStudents - 1) AS "price growth", start.mdate, end.mdate FROM (SELECT DISTINCT name_de FROM dish WHERE restaurant = 'mensa-academica-paderborn' AND category = 'dish') AS ddish JOIN (SELECT name_de, min(date) AS mdate, priceStudents FROM dish WHERE restaurant = 'mensa-academica-paderborn' AND category = 'dish' GROUP BY name_de) AS start ON ddish.name_de = start.name_de JOIN (SELECT name_de, max(date) AS mdate, priceStudents FROM dish WHERE restaurant = 'mensa-academica-paderborn' AND category = 'dish' GROUP BY name_de) AS end ON ddish.name_de = end.name_de AND start.mdate != end.mdate ORDER BY "price growth" DESC

```
